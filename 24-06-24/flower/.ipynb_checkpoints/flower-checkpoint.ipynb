### Load the dataset,Display the first few rows

```python
import pandas as pd

# Data as a string
data = """Sepal Length (cm),Sepal Width (cm),Petal Length (cm),Petal Width (cm),Species
5.1,3.5,1.4,0.2,FlowerA
4.9,3.0,1.4,0.2,FlowerA
4.7,3.2,1.3,0.2,FlowerA
4.6,3.1,1.5,0.2,FlowerA
5.0,3.6,1.4,0.2,FlowerA
7.0,3.2,4.7,1.4,FlowerB
6.4,3.2,4.5,1.5,FlowerB
6.9,3.1,4.9,1.5,FlowerB
5.5,2.3,4.0,1.3,FlowerB
6.5,2.8,4.6,1.5,FlowerB
6.3,3.3,6.0,2.5,FlowerC
5.8,2.7,5.1,1.9,FlowerC
7.1,3.0,5.9,2.1,FlowerC
6.3,2.9,5.6,1.8,FlowerC
6.5,3.0,5.8,2.2,FlowerC"""

# Save to CSV file
with open('sample_dataset.csv', 'w') as file:
    file.write(data)

# Load the data into a DataFrame
df = pd.read_csv('sample_dataset.csv')

# Display the first few rows
print(df.head())
```

#### Generate summary statistics

```python
# Summary statistics
summary = df.describe()

# Mean and standard deviation of Sepal Length
mean_sepal_length = summary.loc['mean', 'Sepal Length (cm)']
std_sepal_length = summary.loc['std', 'Sepal Length (cm)']

print("Summary Statistics:\n", summary)
print("Mean of Sepal Length:", mean_sepal_length)
print("Standard Deviation of Sepal Length:", std_sepal_length)
```

#### Check for Missing Values

```python
# Check for missing values
missing_values = df.isnull().sum()

print("Missing Values:\n", missing_values)

# Handling missing values (if any were found)
# Assuming the strategy would be to fill missing values with the mean of the column
df.fillna(df.mean(), inplace=True)
```

#### Convert Species Labels to Numerical Values

```python
# Mapping dictionary
species_mapping = {'FlowerA': 0, 'FlowerB': 1, 'FlowerC': 2}

# Convert species labels
df['Species'] = df['Species'].map(species_mapping)

print(df.head())
```

#### Split the Dataset

```python
from sklearn.model_selection import train_test_split

# Features and target
X = df.drop('Species', axis=1)
y = df['Species']

# Stratified split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

print("Training Set Size:", len(X_train))
print("Testing Set Size:", len(X_test))
```

####  Train a Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the classifier
clf.fit(X_train, y_train)
```

#### Visualize the Decision Tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=X.columns, class_names=['FlowerA', 'FlowerB', 'FlowerC'], filled=True)
plt.show()
```

#### Predict the Species and Compute Accuracy

```python
from sklearn.metrics import accuracy_score

# Predict the species
y_pred = clf.predict(X_test)

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```

#### Generate a Classification Report and Confusion Matrix

```python
from sklearn.metrics import classification_report, confusion_matrix

# Classification report
report = classification_report(y_test, y_pred, target_names=['FlowerA', 'FlowerB', 'FlowerC'])
print("Classification Report:\n", report)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```
